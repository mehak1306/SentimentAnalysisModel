```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv("tweets.csv")
```

```python
data.head()
```

```python
import re
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

all_sw = stopwords.words('english')
all_sw.remove('not')
```

```python
corpus=[]

for i in range(0, 31962):
  review = re.sub('[^a-zA-Z]', ' ', data['tweet'][i])
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in set(all_sw)]
  review = ' '.join(review)
  corpus.append(review)
```

```python
# Loading BoW dictionary
from sklearn.feature_extraction.text import CountVectorizer
import pickle
cvFile='c1_BoW_Sentiment_Model.pkl'
# cv = CountVectorizer(decode_error="replace", vocabulary=pickle.load(open('./drive/MyDrive/Colab Notebooks/2 Sentiment Analysis (Basic)/3.1 BoW_Sentiment Model.pkl', "rb")))
cv = pickle.load(open(cvFile, "rb"))
```

```python
X_fresh = cv.transform(corpus).toarray()
X_fresh.shape
```

```python
import joblib
classifier = joblib.load('c2_Classifier_Sentiment_Model')
```

```python
y_pred = classifier.predict(X_fresh)
print(y_pred)
```

```python
data['predicted_label'] = y_pred.tolist()
data.head()
```

```python
data.to_csv("c3_Predicted_Sentiments_Fresh_Dump.tsv", sep='\t', encoding='UTF-8', index=False)
```
